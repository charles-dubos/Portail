# Jeu de cartes
comment faire rentrer n cartes sur une dimension fixée.

On a:
- r = L/H le ratio d'une carte
- x, y les dimensions de l'écran (respectivement largeur, hauteur)
- n le nombre de cartes à afficher

On cherche:
- a le nombre de cartes en largeur
- (lié à b le nombre de cartes en hauteur)

```python
import math

n=20
x=16
y=9
r=1

a=math.ceil(math.sqrt(n*x/(y*r)))

print("a=",a,
     "b=", math.ceil(n/a))
```
